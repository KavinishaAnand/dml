```python
# ============================================
# IMPORT LIBRARIES
# ============================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier, GradientBoostingRegressor
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import mean_squared_error, accuracy_score, classification_report, confusion_matrix, roc_auc_score
from sklearn.linear_model import LogisticRegression
from sklearn.cluster import KMeans
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

# Causal Inference Libraries
from econml.dml import LinearDML, CausalForestDML
from econml.metalearners import SLearner, TLearner, XLearner

# Set random seed for reproducibility
np.random.seed(42)

print("✓ All libraries imported successfully!")
print("-" * 80)

# ============================================
# 1. LOAD AND PREPARE DATA
# ============================================

print("\n1. LOADING AND PREPARING DATA...")
print("-" * 80)

# Load your data
df = pd.read_csv("Downloads/dml_synthetic_data.csv")

print(f"✓ Data loaded successfully!")
print(f"  Sample size: {df.shape[0]} rows, {df.shape[1]} columns")
print(f"  Columns: {list(df.columns)}")

# Check if required columns exist
# Assuming your CSV has columns like: treatment, outcome, and features
# You may need to adjust these column names based on your actual data

# Display basic information about the data
print(f"\n✓ Data Overview:")
print(f"  First few rows:")
print(df.head())
print(f"\n  Data types:")
print(df.dtypes)
print(f"\n  Missing values:")
print(df.isnull().sum())

# ============================================
# 2. DATA PREPROCESSING
# ============================================

print("\n2. DATA PREPROCESSING...")
print("-" * 80)

# Identify treatment, outcome, and feature columns
# You'll need to adjust these based on your actual column names
# Common column names might be: 'treatment', 'outcome', 'y', 't', etc.

# Let's try to automatically identify key columns
def identify_columns(df):
    """Try to automatically identify treatment, outcome, and feature columns."""
    treatment_cols = [col for col in df.columns if 'treatment' in col.lower() or 't' == col.lower() or 'treat' in col.lower()]
    outcome_cols = [col for col in df.columns if 'outcome' in col.lower() or 'y' == col.lower() or 'effect' in col.lower()]
    
    treatment_col = treatment_cols[0] if treatment_cols else None
    outcome_col = outcome_cols[0] if outcome_cols else None
    
    # Features are all other columns
    feature_cols = [col for col in df.columns if col not in [treatment_col, outcome_col]]
    
    return treatment_col, outcome_col, feature_cols

treatment_col, outcome_col, feature_cols = identify_columns(df)

# If automatic identification failed, you may need to set these manually:
# treatment_col = 'treatment'  # Adjust based on your column name
# outcome_col = 'outcome'      # Adjust based on your column name
# feature_cols = [col for col in df.columns if col not in [treatment_col, outcome_col]]

print(f"✓ Identified columns:")
print(f"  Treatment column: {treatment_col}")
print(f"  Outcome column: {outcome_col}")
print(f"  Feature columns: {feature_cols}")

# Check if we found the necessary columns
if treatment_col is None or outcome_col is None:
    print("\n⚠ WARNING: Could not automatically identify treatment or outcome columns.")
    print("Please set them manually in the code.")
    print("Available columns:", list(df.columns))
    # You'll need to uncomment and adjust these lines:
    # treatment_col = 'your_treatment_column_name'
    # outcome_col = 'your_outcome_column_name'
    # feature_cols = [col for col in df.columns if col not in [treatment_col, outcome_col]]
else:
    # Define X, T, Y
    X = df[feature_cols]
    T = df[treatment_col]
    Y = df[outcome_col]
    
    # Check treatment distribution
    print(f"\n✓ Treatment distribution:")
    print(f"  Treatment rate: {T.mean():.2%}")
    print(f"  Number treated: {T.sum()}")
    print(f"  Number control: {len(T) - T.sum()}")
    
    # Check outcome statistics
    print(f"\n✓ Outcome statistics:")
    print(f"  Overall mean: {Y.mean():.4f}")
    print(f"  Treated mean: {Y[T == 1].mean():.4f}")
    print(f"  Control mean: {Y[T == 0].mean():.4f}")
    print(f"  Difference: {Y[T == 1].mean() - Y[T == 0].mean():.4f}")

# ============================================
# 3. DATA SPLITTING
# ============================================

print("\n3. SPLITTING DATA INTO TRAIN/TEST SETS...")
print("-" * 80)

# Split the data
X_train, X_test, Y_train, Y_test, T_train, T_test = train_test_split(
    X, Y, T, test_size=0.2, random_state=42, stratify=T
)

print(f"✓ Data split completed!")
print(f"  Training set: {X_train.shape[0]} samples")
print(f"  Test set: {X_test.shape[0]} samples")
print(f"  Treatment rate (train): {T_train.mean():.2%}")
print(f"  Treatment rate (test): {T_test.mean():.2%}")

# ============================================
# 4. PROPENSITY SCORE MODELING
# ============================================

print("\n4. PROPENSITY SCORE MODELING...")
print("-" * 80)

# Train propensity model
propensity_model = LogisticRegression(random_state=42, max_iter=1000)
propensity_model.fit(X_train, T_train)

# Predict propensity scores
propensity_train = propensity_model.predict_proba(X_train)[:, 1]
propensity_test = propensity_model.predict_proba(X_test)[:, 1]

# Calculate accuracy
T_pred = propensity_model.predict(X_test)
propensity_accuracy = accuracy_score(T_test, T_pred)

# Check overlap
overlap_percentage = np.mean((propensity_test > 0.1) & (propensity_test < 0.9)) * 100

print(f"✓ Propensity model accuracy: {propensity_accuracy:.3f}")
print(f"✓ Propensity score range: [{propensity_test.min():.3f}, {propensity_test.max():.3f}]")
print(f"✓ Overlap check: {overlap_percentage:.1f}% in [0.1, 0.9]")

# ============================================
# 5. BALANCE CHECK (OPTIONAL)
# ============================================

print("\n5. BALANCE CHECK...")
print("-" * 80)

# Calculate standardized mean differences
def calculate_smd(feature, treatment, propensity):
    treated = feature[treatment == 1]
    control = feature[treatment == 0]
    
    mean_diff = treated.mean() - control.mean()
    pooled_std = np.sqrt((treated.std()**2 + control.std()**2) / 2)
    
    return abs(mean_diff / pooled_std) if pooled_std > 0 else 0

smds = {}
for col in X.columns:
    smds[col] = calculate_smd(X_test[col], T_test, propensity_test)

print("✓ Standardized Mean Differences (SMD):")
for col, smd in smds.items():
    status = "✓" if smd < 0.1 else "⚠"
    print(f"  {status} {col}: {smd:.4f}")

# ============================================
# 6. OUTCOME MODELING (BASELINE)
# ============================================

print("\n6. BASELINE OUTCOME MODELING...")
print("-" * 80)

# Simple outcome model
outcome_model = RandomForestRegressor(n_estimators=100, random_state=42)
outcome_model.fit(X_train, Y_train)

# Predictions
Y_pred_train = outcome_model.predict(X_train)
Y_pred_test = outcome_model.predict(X_test)

# Calculate RMSE
train_rmse = np.sqrt(mean_squared_error(Y_train, Y_pred_train))
test_rmse = np.sqrt(mean_squared_error(Y_test, Y_pred_test))

print(f"✓ Outcome model trained successfully!")
print(f"  Training RMSE: {train_rmse:.4f}")
print(f"  Test RMSE: {test_rmse:.4f}")

# ============================================
# 7. CAUSAL INFERENCE METHODS
# ============================================

print("\n" + "="*80)
print("7. CAUSAL INFERENCE METHODS")
print("="*80)

# Initialize results dictionary
results = {}

# 7.1 Linear Double Machine Learning (DML)
print("\n7.1 Linear Double Machine Learning (DML)...")
try:
    linear_dml = LinearDML(
        model_y=RandomForestRegressor(n_estimators=100, random_state=42),
        model_t=RandomForestClassifier(n_estimators=100, random_state=42),
        discrete_treatment=True,
        random_state=42
    )
    linear_dml.fit(Y_train, T_train, X=X_train)
    ate_linear = linear_dml.ate(X=X_test)
    results['Linear DML'] = {
        'ATE': ate_linear.mean(), 
        'CI': (
            ate_linear.mean() - 1.96 * ate_linear.std(),
            ate_linear.mean() + 1.96 * ate_linear.std()
        )
    }
    print(f"✓ Linear DML ATE: {ate_linear.mean():.4f}")
except Exception as e:
    print(f"✗ Linear DML failed: {e}")

# 7.2 Causal Forest DML
print("\n7.2 Causal Forest DML...")
try:
    causal_forest = CausalForestDML(
        model_y=RandomForestRegressor(n_estimators=100, random_state=42),
        model_t=RandomForestClassifier(n_estimators=100, random_state=42),
        discrete_treatment=True,
        n_estimators=100,
        random_state=42
    )
    causal_forest.fit(Y_train, T_train, X=X_train)
    ate_forest = causal_forest.ate(X=X_test)
    results['Causal Forest'] = {
        'ATE': ate_forest.mean(),
        'CI': (
            ate_forest.mean() - 1.96 * ate_forest.std(),
            ate_forest.mean() + 1.96 * ate_forest.std()
        )
    }
    print(f"✓ Causal Forest ATE: {ate_forest.mean():.4f}")
except Exception as e:
    print(f"✗ Causal Forest failed: {e}")

# 7.3 Meta-Learners
print("\n7.3 Meta-Learners...")

# S-Learner
print("  - S-Learner...")
try:
    s_learner = SLearner(overall_model=RandomForestRegressor(n_estimators=100, random_state=42))
    s_learner.fit(Y_train, T_train, X=X_train)
    ate_s = s_learner.estimate_ate(X=X_test)
    results['S-Learner'] = {
        'ATE': ate_s[0][0],
        'CI': (ate_s[1][0][0], ate_s[1][1][0]) if len(ate_s) > 1 else (None, None)
    }
    print(f"    ✓ S-Learner ATE: {ate_s[0][0]:.4f}")
except Exception as e:
    print(f"    ✗ S-Learner failed: {e}")

# T-Learner
print("  - T-Learner...")
try:
    t_learner = TLearner(
        models=RandomForestRegressor(n_estimators=100, random_state=42)
    )
    t_learner.fit(Y_train, T_train, X=X_train)
    ate_t = t_learner.estimate_ate(X=X_test)
    results['T-Learner'] = {
        'ATE': ate_t[0][0],
        'CI': (ate_t[1][0][0], ate_t[1][1][0]) if len(ate_t) > 1 else (None, None)
    }
    print(f"    ✓ T-Learner ATE: {ate_t[0][0]:.4f}")
except Exception as e:
    print(f"    ✗ T-Learner failed: {e}")

# X-Learner
print("  - X-Learner...")
try:
    x_learner = XLearner(
        models=RandomForestRegressor(n_estimators=100, random_state=42),
        propensity_model=propensity_model
    )
    x_learner.fit(Y_train, T_train, X=X_train)
    ate_x = x_learner.estimate_ate(X=X_test)
    results['X-Learner'] = {
        'ATE': ate_x[0][0],
        'CI': (ate_x[1][0][0], ate_x[1][1][0]) if len(ate_x) > 1 else (None, None)
    }
    print(f"    ✓ X-Learner ATE: {ate_x[0][0]:.4f}")
except Exception as e:
    print(f"    ✗ X-Learner failed: {e}")

# ============================================
# 8. RESULTS SUMMARY
# ============================================

print("\n" + "="*80)
print("8. RESULTS SUMMARY")
print("="*80)

print("\nAverage Treatment Effects (ATE):")
print("-" * 40)
for method, result in results.items():
    if 'ATE' in result and result['ATE'] is not None:
        ci = result['CI']
        ci_str = f"[{ci[0]:.4f}, {ci[1]:.4f}]" if ci[0] is not None else "Not available"
        print(f"{method:20} ATE: {result['ATE']:8.4f}  95% CI: {ci_str}")

# ============================================
# 9. VISUALIZATION (OPTIONAL)
# ============================================

print("\n" + "="*80)
print("9. VISUALIZATIONS")
print("="*80)

try:
    # Plot ATE comparison
    fig, axes = plt.subplots(1, 2, figsize=(15, 5))
    
    # Plot 1: ATE comparison
    methods = []
    ate_values = []
    ate_errors = []
    
    for method, result in results.items():
        if 'ATE' in result and result['ATE'] is not None and result['CI'][0] is not None:
            methods.append(method)
            ate_values.append(result['ATE'])
            ci_lower, ci_upper = result['CI']
            ate_errors.append([result['ATE'] - ci_lower, ci_upper - result['ATE']])
    
    if methods:
        axes[0].errorbar(methods, ate_values, yerr=np.array(ate_errors).T, 
                        fmt='o', capsize=5, markersize=8)
        axes[0].axhline(y=0, color='r', linestyle='--', alpha=0.5)
        axes[0].set_xlabel('Method')
        axes[0].set_ylabel('ATE')
        axes[0].set_title('Average Treatment Effect Comparison')
        axes[0].tick_params(axis='x', rotation=45)
    
    # Plot 2: Propensity score distribution
    axes[1].hist(propensity_test[T_test == 0], alpha=0.5, label='Control', bins=20)
    axes[1].hist(propensity_test[T_test == 1], alpha=0.5, label='Treated', bins=20)
    axes[1].set_xlabel('Propensity Score')
    axes[1].set_ylabel('Frequency')
    axes[1].set_title('Propensity Score Distribution')
    axes[1].legend()
    axes[1].axvline(x=0.1, color='r', linestyle='--', alpha=0.5)
    axes[1].axvline(x=0.9, color='r', linestyle='--', alpha=0.5)
    
    plt.tight_layout()
    plt.savefig('causal_analysis_results.png', dpi=300, bbox_inches='tight')
    plt.show()
    print("✓ Visualizations saved to 'causal_analysis_results.png'")
    
except Exception as e:
    print(f"✗ Visualization failed: {e}")

# ============================================
# 10. SAVE RESULTS
# ============================================

print("\n" + "="*80)
print("10. SAVING RESULTS")
print("="*80)

try:
    # Save results to CSV
    results_df = pd.DataFrame(results).T
    results_df.to_csv('causal_analysis_results.csv')
    print("✓ Results saved to 'causal_analysis_results.csv'")
    
    # Save propensity scores
    propensity_df = pd.DataFrame({
        'propensity_score': propensity_test,
        'treatment': T_test,
        'outcome': Y_test
    })
    propensity_df.to_csv('propensity_scores.csv', index=False)
    print("✓ Propensity scores saved to 'propensity_scores.csv'")
    
    # Save dataset summary
    summary = {
        'dataset': 'dml_synthetic_data.csv',
        'n_samples': len(df),
        'treatment_rate': T.mean(),
        'outcome_mean': Y.mean(),
        'n_features': len(feature_cols)
    }
    summary_df = pd.DataFrame([summary])
    summary_df.to_csv('dataset_summary.csv', index=False)
    print("✓ Dataset summary saved to 'dataset_summary.csv'")
    
except Exception as e:
    print(f"✗ Saving results failed: {e}")

# ============================================
# 11. FINAL SUMMARY
# ============================================

print("\n" + "="*80)
print("11. EXECUTIVE SUMMARY")
print("="*80)

print("\n" + "✓" * 80)
print("✓ All analyses completed successfully!")
print(f"✓ Dataset: dml_synthetic_data.csv ({len(df)} samples)")
print(f"✓ Treatment rate: {T.mean():.2%}")
print(f"✓ Number of causal methods run: {len(results)}")
print("✓ Results saved to files")
print("✓ Visualizations generated")
print("\n✓ Ready for decision-making and implementation.")
print("✓" * 80)
```
